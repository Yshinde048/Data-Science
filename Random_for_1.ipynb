```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
company=pd.get_dummies(company,columns=['ShelveLoc','US','Urban'])
company
```

```python
company['Income'].describe()
```

```python
lab=pd.cut(company.Sales,bins=[0,5.39,9.32,17],labels=['low','medium','high'])
company.insert(0,'Income_label',lab)
```

```python
company=company.drop('Income',axis=1)
company
```

```python
company['Income_label'].isna().sum()
```

```python
company['Income_label']=company['Income_label'].fillna('low')
```

```python
company['Income_label'].isna().sum()
```

```python
y=company.iloc[:,0]
x=company.iloc[:,1:]
y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestClassifier

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=90)
```

```python
kfold=KFold(n_splits=5)
model1=RandomForestClassifier(n_estimators=200,max_features=3)
model1.fit(x_train,y_train)
```

```python
pred=model1.predict(x_test)
pred
```

```python
pd.crosstab(pred,y_test)
```
